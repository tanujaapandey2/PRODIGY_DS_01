## Finding top 15 countries with respect to total population

```python
#filter data for total population
total_population_data=df[df['Series Code']=='SP.POP.TOTL']
#SORT DATA BASED ON THE TOTAL POPULATION FOR 2022
total_population_sorted=total_population_data.sort_values(by='2022',ascending=False)
#top population for2022
total_top_fiften_countries=total_population_sorted.head(15)
print(total_top_fiften_countries[['Country Code']])
```

## Bar plot

# top ten countries total population in year 2022 and 2016

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2022",y="Country Code",data=total_top_fiften_countries)
plt.title("Top fiften countries of total population(2022)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2016",y="Country Code",data=total_top_fiften_countries)
plt.title("Top fiften countries of total population(2016)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()


```

## Finding Bottom 15 countries with respect to total population

```python
#filter data for total population
total_population_data=df[df['Series Code']=='SP.POP.TOTL']
#SORT DATA BASED ON THE TOTAL POPULATION FOR 2022
total_population_sorted1=total_population_data.sort_values(by='2022',ascending=True)
#bottom population for2022
total_bottom_fiften_countries=total_population_sorted1.tail(15)
print(total_bottom_fiften_countries[['Country Code']])
```

# top bottom countries total population in year 2022 and 2016

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2022",y="Country Code",data=total_bottom_fiften_countries)
plt.title("Bottom fiften countries of total population(2022)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2016",y="Country Code",data=total_bottom_fiften_countries)
plt.title("Bottom fiften countries of total population(2016)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
#Finding top 15 Countries with highest male population
male_population_data = df[df['Series Code']=='SP.POP.TOTL']
male_population_sorted=male_population_data.sort_values(by='2022',ascending=False)
male_top_fiften_countries=male_population_sorted.head(15)
print("top fiften male population")
print(male_top_fiften_countries[['Country Code']])
```

```python
female_population_data = df[df['Series Code']=='SP.POP.TOTL']
female_population_sorted=female_population_data.sort_values(by='2022',ascending=False)
female_top_fiften_countries=female_population_sorted.head(15)
print("top fiften female population")
print(female_top_fiften_countries[['Country Code']])
```

## Top fiften countries with highest male and fimale population in 2022

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2022",y="Country Code",data=male_top_fiften_countries)
plt.title("Top f fiften countries of Male population(2022)",fontsize=10)
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
snas.barplot(x="2016",y="Country Code",data=female_top_fiften_countries)
plt.title("top fiften countries of female population(2016)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
## Top 10 Countriesof male and female by STACKED BAR PLOT
```

```python
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
merge_data["Total population"]=merge_data["2022_male"]+merge_data["2022_female"]
```

```python
merge_data.head()
```

```python
merge_data.tail()
```

## Select the Top 15 Countries with thehighest total population

```python
sorted_data=merge_data.sort_values(by="Total population",ascending=False)
```

```python
top_15_countries=sorted_data.head(15)
```

```python
plt.figure(figsize=(12,6))
snas.barplot(x="Country Code",y="2022_female",data=top_15_countries,color="red",label="female population")
snas.barplot(x="Country Code",y="2022_male",data=top_15_countries,bottom=top_15_countries["2022_female"],color="blue",label="male population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.show
```

## Bottom 15 countries with male and female 2022

```python
bottom_15_countries=sorted_data.tail(15)
```

```python
plt.figure(figsize=(12,6))
snas.barplot(x="Country Code",y="2022_female",data=bottom_15_countries,color="red",label="female population")
snas.barplot(x="Country Code",y="2022_male",data=bottom_15_countries,bottom=bottom_15_countries["2022_female"],color="blue",label="male population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.show
```
